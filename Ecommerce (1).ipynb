```python
## checking for missing values 		customer_id	first_name	last_name	gender	age	state
pd.read_sql("""select 
            SUM(case when customer_id is null then 1 else 0 end) as null_customer_id,
            SUM(case when first_name is null then 1 else 0 end) as null_customer_name,
            SUM(case when  Last_name is null then 1 else 0 end) as null_customer_lastname,
            SUM(case when gender is null then 1 else 0 end) as null_customer_gender,
            SUM(case when age is null then 1 else 0 end) as null_customer_age,
            SUM(case when state is null then 1 else 0 end) as null_customer_state
            FROM customers;""", conn)
```

```python
# CHECKING FOR MISSING VALUES IN PRODUCT 	product_id	product_name	category	price
pd.read_sql(""" select
                 sum(case when product_id is null then 1 else 0 end) as null_product_id,
                 sum(case when product_name is null then 1 else 0 end) as null_product_name,
                 sum(case when category is null then 1 else 0 end) as null_product_category,
                 sum(case when price is null then 1 else 0 end) as null_product_price
                 from products; """, conn)
```

```python
## checking for missing values in orders order_id	customer_id	order_date	shipped_date	status
pd.read_sql(""" select
                 sum(case when order_id is null then 1 else 0 end) as null_order_id,
                 sum(case when customer_id is null then 1 else 0 end) as null_customer_id,
                 sum(case when order_date is null then 1 else 0 end) as null_orders_date,
                 sum(case when shipped_date is null then 1 else 0 end) as null_shipped_date,
                 sum(case when status is null then 1 else 0 end) as null_status
                 from orders;""", conn)
```

```python
## checking for missing value in order_item  order_id	product_id	quantity	price
pd.read_sql(""" select
                     sum(case when price is null then 1 else 0 end) as null_price,
                  sum(case when order_id is null then 1 else 0 end) as null_order_id,
                  sum(case when product_id is null then 1 else 0 end) as null_product_id,
                  sum(case when quantity is null then 1 else 0 end) as null_quantity
                  from order_items;""", conn)
```

```python


query = """
SELECT 
    customers.customer_id,
    customers.first_name, 
    order_items.quantity,
    orders.order_id,
    SUM(order_items.quantity * products.price),
    sum(order_items.price) AS total_spend,
    count(distinct orders.order_id) as total_orders
FROM orders
JOIN order_items ON orders.order_id = order_items.order_id
JOIN customers ON orders.customer_id = customers.customer_id
JOIN products ON order_items.product_id = products.product_id
GROUP BY customers.customer_id, customers.first_name
ORDER BY total_spend DESC
;
"""

top_customers = pd.read_sql_query(query, conn)
top_customers
```

```python
## What is the average order value per order?
```

```python
query = """
SELECT 
    ROUND(AVG(order_value), 2) AS average_order_value
FROM (
    SELECT 
        orders.order_id,
        SUM(order_items.quantity * products.price) AS order_value
    FROM orders
    JOIN order_items ON orders.order_id = order_items.order_id
    JOIN products ON order_items.product_id = products.product_id
    GROUP BY orders.order_id
)
"""
avg_orders = pd.read_sql_query(query, conn)
avg_orders
```

```python
### calculating the total amount per order

query = """SELECT 
    customers.customer_id,
    customers.first_name,
    orders.order_id,
    ROUND(SUM(order_items.quantity * products.price), 2) AS order_value
FROM customers
JOIN orders ON customers.customer_id = orders.customer_id
join order_items on orders.order_id = order_items.order_id
JOIN products ON order_items.product_id = products.product_id
GROUP BY  customers.customer_id,
    customers.first_name,
    orders.order_id 
;"""

avg_orders= pd.read_sql_query(query, conn)
avg_orders
```

```python
# Calculating the average order value for each customer
df = pd.read_sql_query(query, conn)

avg_orders = df.groupby(['customer_id',  'first_name'])['order_value'].mean().reset_index()

# Renaming the column for clarity
avg_orders.rename(columns={'order_value': 'avg_orders'}, inplace=True)

avg_orders
```

```python
## sorting the avg values
avg_orders.sort_values(by='avg_orders', ascending=False, inplace=True)
avg_orders
```

```python
## customer distribution by location
query = """ select
  customers.state as location,
  count(distinct customers.customer_id) as count
  from 
  customers
  group by location
  order by  count """

customer_location = pd.read_sql_query(query, conn)
customer_location 
```

```python
## customer shoping frequency
query = """ Select
               customers.first_name, customers.state,
               orders.order_id,
               count(distinct orders.order_id) as total_orders
               FROM orders
JOIN order_items ON orders.order_id = order_items.order_id
JOIN customers ON orders.customer_id = customers.customer_id
JOIN products ON order_items.product_id = products.product_id
GROUP BY customers.state, customers.first_name
ORDER BY total_orders Desc; """

shopping_frequency = pd.read_sql_query(query,conn)
shopping_frequency              
```
